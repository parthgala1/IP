```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("flower.jpg")
image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(image, cmap='gray')
plt.axis("off")
```

```python
import numpy as np

def contrast_stretching(image, a, b, l, m, n):
    image = image.astype(np.float32)
    v = l * a
    image4 = image.copy()
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]): 
            if image4[i, j] <= a:
                image4[i, j] = l * image4[i, j]
            elif a < image4[i, j] <= b:
                image4[i, j] = m * (image4[i, j] - a) + v
            elif image4[i, j] > b:
                image4[i, j] = n * (image4[i, j] - b) + m * (b - a) + v

    return np.clip(image4, 0, 255).astype(np.uint8)
```

```python
image2 = contrast_stretching(image, 50, 200, 0.1, 1.2, 0.2)
plt.imshow(image2, cmap='gray')
```
