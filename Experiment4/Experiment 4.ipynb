```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('spiderman.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
filter = np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])
```

```python
def high_pass_filter(img, filter):
    image = img.copy()
    for i in range(1, img.shape[0]-1):
        for j in range(1, img.shape[1]-1):
            image[i][j] = np.sum(img[i-1:i+2, j-1:j+2] * filter)
    return image
```

```python
image = high_pass_filter(img, filter)
plt.imshow(image, cmap='gray')
plt.axis('off')
```

```python
def high_boost_filter(img, filter, alpha):
    image = img.copy()
    for i in range(1, img.shape[0]-1):
        for j in range(1, img.shape[1]-1):
            image[i][j] = np.sum(img[i-1:i+2, j-1:j+2] * filter) + alpha * img[i][j]

    image = np.clip(image, 0, 255)
    return image
```

```python
filter = np.array([[-1, -1, -1],
                    [-1, 8.9, -1],
                    [-1, -1, -1]])

image = high_boost_filter(img, filter, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')
```
