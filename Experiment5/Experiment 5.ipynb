```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.axis("off")
```

```python
s_min = 0
s_max = 255
r_min = np.min(img)
r_max = np.max(img)
```

```python
r_min, r_max
```

```python
def histogram_map(img):
  height = img.shape[0]
  width = img.shape[1]
  img_hist = list(range(0, 256))
  r = []
  for i in range(height):
    for j in range(width):
      img_hist[img[i, j]]+=1

  return (img_hist)
```

```python
count1 = histogram_map(img)

plt.plot(count1, color='c')
plt.xlabel('intensity value')
plt.ylabel('number of pixels')
plt.title('Histogram of the original image')
```

```python
def stretching(img):
  height = img.shape[0]
  width = img.shape[1]
  for i in range(height):
    for j in range(width):
      stretched_img = ((s_max - s_min)/(r_max - r_min))*(img[i, j] - r_min) + s_min
      img[i, j] = stretched_img
  stretched_img = np.clip(img, 0, 255)
  return img
```

```python
new_img = stretching(img)
```

```python
plt.imshow(new_img, cmap='gray')
plt.axis('off')
```

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
count2 = histogram_map(new_img)

plt.plot(count1, color="blue")
plt.plot(count2, color="yellow")
plt.legend()
plt.xlabel('intensity value')
plt.ylabel('number of pixels')
plt.title('Histogram of the image')
```
